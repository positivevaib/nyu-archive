```python
import matplotlib.pyplot as plt
```

```python
mbw = 30
pflops = 200
cross = pflops/mbw
```

```python
ai = [0, 1/8., 1/4., 1/2., 2, 4, 6, 8, 10, 32, cross]
ai.sort()
ai
```

```python
flops = [min(pflops, mbw*i) for i in ai]
flops
```

```python
plt.plot(20, 20)
plt.grid()
plt.plot(ai[:8], flops[:8], '-b', linewidth=4)
plt.plot(ai[7:], [mbw*i for i in ai[7:]], ':b')
plt.plot(ai[7:], flops[7:], '-r', linewidth=4)
plt.plot([0, cross], [pflops, pflops], ':r')
plt.plot([cross, cross], [0, pflops], ':g')
plt.xscale('log', basex=2)
plt.yscale('log', basey=2)
plt.xlabel('Arith. Intensity [FLOP/bytes]  log-scale')
plt.ylabel('Actual FLOPS  log-scale')
plt.text(1.5, 0.25, 'Memory Bound', color='green')
plt.text(cross + 1, 0.25, 'Compute Bound', color='green')
plt.text(0.25, 275, 'Peak FLOPS limit', color='red')
plt.text(0.12, 6, 'DRAM bandwidth limit [GB/sec.]', color='blue', rotation=16)

dp11, = plt.plot(2.012316/8.049263, 2.012316, '.', color='orange', label='dp1 - 1000000')
dp12, = plt.plot(1.673381/6.693523, 1.673381, '^', color='orange', label='dp1 - 300000000')

dp21, = plt.plot(4.726070/18.904282, 4.726070, '.', color='darkgreen', label='dp2 - 1000000')
dp22, = plt.plot(2.576099/10.304395, 2.576099, '^', color='darkgreen', label='dp2 - 300000000')

dp31, = plt.plot(7.664893/30.659572, 7.664893, '.', color='cyan', label='dp3 - 1000000')
dp32, = plt.plot(2.877008/11.508031, 2.877008, '^', color='cyan', label='dp3 - 300000000')

dp41, = plt.plot(0.004226/0.016906, 0.004226, '.', color='purple', label='dp4 - 1000000')
dp42, = plt.plot(0.004207/0.016829, 0.004207, '^', color='purple', label='dp4 - 300000000')

dp51, = plt.plot(7.365132/29.460529, 7.365132, '.', color='magenta', label='dp5 - 1000000')
dp52, = plt.plot(2.868224/11.472895, 2.868224, '^', color='magenta', label='dp5 - 300000000')

plt.legend(loc='best', bbox_to_anchor=(1, 0.85), ncol=1)
plt.title('Roofline Model - 200 GFLOPS and 30 GB/sec. ')
#plt.show()
plt.savefig('roof.pdf', bbox_inches='tight')
```
